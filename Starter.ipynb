```python
openai.api_key = ""
temperature_set = 0.0
model_name = "gpt-3.5-turbo-0125"  # "gpt-3.5-turbo-0125" #"gpt-3.5-turbo-16k"# "gpt-4"
```

```python
t1 = '''You are an expert in carbon emissions analysis, tasked with estimating the carbon footprint of an individual based on their daily behaviors. Your goal is to provide a detailed breakdown of the user's carbon emissions or reductions across various categories of their lifestyle.

Here is the user's description of their daily behaviors:'''


t2 = '''Please analyze this input and provide a structured assessment of the user's carbon footprint. Follow these steps:

1. Carefully read the user's input and identify distinct behaviors that may impact their carbon footprint.

2. For each identified behavior:
   a. Categorize it (e.g., Transportation, Home Energy, Gardening, etc.)
   b. Estimate its carbon emission or reduction range per month in kilograms (kg)
   c. Assign a descriptive name to the behavior
   d. Note any additional relevant information provided by the user

3. Compile your analysis into a JSON object following this structure:
   {
     "items": [
       {
         "category": "Category name",
         "emission_range": "X-Y kg",
         "name": "Descriptive name of the behavior",
         "additional_info": "Any extra details provided"
       },
       // Additional items as needed
     ]
   }

4. If the user is providing informationa bout members of his family or multiple people in his home, then provide ALL the emission ranges cumulatively for all members. 
Stay consistent in your emission estimates - of whether you are doing it for a single person or multiple people. 

4. Based on the emission values and ranges, provide a final assessment of the user (or the total of the user's family/multiple members depending on the case) carbon footprint in the format: \" Estimated monthly carbon footprint: \"x\" kg \" .
It should be a single numerical value in kilograms (kg) representing the estimated monthly carbon emissions. Strictly stick to the format mentioned above for the final estimate.
If you are providing a range, please provide the average of the range as the final estimate.

Note:
- Emission ranges except the final estimate should be given as "X - Y kg" where X and Y are numbers.
- Negative emission ranges (e.g., "-5 - -1 kg") indicate carbon reduction activities.
- Be as precise as possible with your estimates, but understand that these are approximations.


Provide the final JSON output as described above with the estimated monthly carbon emissions. '''
```

```python
user_input = ''' I drive a 2013 Honda accord 15 miles to and from my work to my house. I have a family of 4 with my wife and 2 kids, all of us are non-vegetarians. We have a double door refrigerator and we live in Atlanta in a 2b2b appartment. I keep my thermostat around 68° fahrenheit in summers and 76 degrees fahrenheit in winters . We use all basic appliances like dishwasher, toaster, HVAC, and we have 4 laptops, phones and other small gadgets. In my small balcony which gets ample sunlight, I have 3-4 potted plants.  All the bulbs in my house are LED bulbs. '''
```

```python
prompt = t1 + '\n\n ##### \n' \
    + user_input + "\n\n" \
    + t2
```

```python
# Create OpenAI client and get response
client = openai.Client(api_key=openai.api_key)

response = client.chat.completions.create(
            model=model_name,
            messages=[{"role": "user", "content": prompt}],
            temperature=temperature_set,
            max_tokens=2000,
            )
```
