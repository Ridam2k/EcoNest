```python
def format_emissions(data):
    output = "User's carbon emissions are:\n"
    for i, item in enumerate(data['items'], 1):
        output += f"{i}. {item['category']}: ({item['emission_range']}, {item['name']})\n"
    return output

emission_data = format_emissions(parsed_data)
```
